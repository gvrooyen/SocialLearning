The first reference technique just plays random moves(Random.py); the second technique is the one given as example in the tournament documentation(Reference.py).

Reference agent: model bias---------------------------We start by exploring the behaviour of the reference agent (Reference.py) given in the tournamentdocumentation. In the first analysis, we plot the performance of the simples case (non-spatial,non-cumulative and without model bias) against the case where model bias is enabled.

```python
fig = plt.figure()plotfitness.axes = Axes3D(fig)filter = {'param_mode_spatial': False,           'param_mode_cumulative': False,           'param_mode_model_bias': False,           'param_N_observe': 5,           'agent_name': 'Reference.py'}              plotfitness.surfplot(dbc, 'Reference.py', filter, 'hot')filter = {'param_mode_spatial': False,           'param_mode_cumulative': False,           'param_mode_model_bias': True,           'param_N_observe': 5,           'agent_name': 'Reference.py'}    plotfitness.surfplot(dbc, 'Reference.py', filter, 'cool')
```

The model bias case is completely outperformed by the basic case where a random model is chosen. Thisseems to suggest that (at least for the reference strategy) it is more valuable to allow a broader segmentof the population to explore payoffs and act as models, rather than allowing only an older subset of thepopulation to act as models.Reference agent: Cumulative mode--------------------------------Next, we compare the base case with the case where the REFINE move is allowed.

```python
fig = plt.figure()plotfitness.axes = Axes3D(fig)filter = {'param_mode_spatial': False,           'param_mode_cumulative': False,           'param_mode_model_bias': False,           'param_N_observe': 5,           'agent_name': 'Reference.py'}              plotfitness.surfplot(dbc, 'Reference.py', filter, 'hot')filter = {'param_mode_spatial': False,           'param_mode_cumulative': True,           'param_mode_model_bias': False,           'param_N_observe': 5,           'agent_name': 'Reference.py'}    plotfitness.surfplot(dbc, 'Reference.py', filter, 'cool')
```

Noticable here is that the strategy seems to benefit from high values of $P_c$ (probability thatan act's payoff changes). In the reference strategy, this triggers another OBSERVE move, whichseems to allow high-value acts to propagate better.

```python
dbc = db.trace_payoffsestimate = []X = range(0,10000,100)for (i, record) in enumerate(dbc.find(limit=100)):    trace = 1. * array(record['trace'])    trace /= trace[-1]    trace = trace[0::100]    trace = abs(trace - 1.)    diff = zeros(100)    diff[1:100] = trace[1:100] - trace[0:99]    estimate.append(trace)trace_mean = mean(estimate,0)trace_std = std(estimate,0)plt.figure()plt.errorbar(X,trace_mean,trace_std)plt.xlabel('Round')plt.ylabel('Error in payoff estimate')plt.figure()plt.plot(X,diff)plt.xlabel('Round')plt.ylabel('Rate of change of payoff estimate')
```

The upper graph shows how close the average payoff per round at the specified turn (measured from thefirst round) is to the final average payoff per round. The estimate is 10 % accurate after less than1000 rounds, but then gradually improves until the final round (the estimate can likely be improved bymeasuring the payoff over the last, say, 500 rounds [moving window] rather than from the firsteach time). The rate of improvement in the estimate, however, (shown in the lower graph) indicates thatthe estimate improves linearly after about round 1000.It can be concluded that, for the purposes of rapid comparison of strategies (e.g. during evolutionaryprogramming), fitness can be roughly estimated after about 1000 rounds, rather than the full 10,000 ofthe tournament.
